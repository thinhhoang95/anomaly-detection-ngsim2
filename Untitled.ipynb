```python
import pandas as pd
from ngsim_processor import NGSIM_Processor as NGSIM

if __name__ == '__main__':
    container_t = []
    container_x = []
    container_y = []
    num_of_trajs = 50
    max_length_t = 0

    df = pd.read_csv('kalmaned2.csv')
    ngsim = NGSIM(df)
    ngsim.set_patch(0.0, 75.0, 500.0, 750.0) # see the notebook df_stats.py to see where these numbers come from

    for i in range(num_of_trajs):
        t, x, y = ngsim.get_trajectory_random()
        container_t.append(t)
        container_x.append(x)
        container_y.append(y)
        if len(t) > max_length_t:
            max_length_t = len(t)
        print('Trajectory {:d} generated'.format(i))
```

```python
from matplotlib import pyplot as plt
```

```python
tx = np.zeros((num_of_trajs, max_length_t))
xx = np.zeros((num_of_trajs, max_length_t))
yx = np.zeros((num_of_trajs, max_length_t))
```

```python
for i in range(num_of_trajs):
    tx[i,0:len(container_t[i])] = container_t[i] - container_t[i][0]
    xx[i,0:len(container_t[i])] = container_x[i]
    yx[i,0:len(container_t[i])] = container_y[i]
    plt.plot(tx[i,:],yx[i,:])
```

```python
for i in range(num_of_trajs):
    plt.plot(tx[i,:],xx[i,:])
```
