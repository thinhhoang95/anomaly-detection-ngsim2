```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('kalmaned2.csv')
df['Global_Time'] = df['Global_Time'] - df['Global_Time'].min()
df.head()
```

```python
veh_665 = df[(df['Vehicle_ID'] == 665) & (df['Local_Y'] > 517)]['Local_Y'].to_numpy()
veh_658 = df[(df['Vehicle_ID'] == 658) & (df['Local_Y'] > 579)]['Local_Y'].to_numpy()
```

```python
plt.plot(veh_665)
plt.plot(veh_658)
```
