## Functional PCA Decomposition

To reduce number of dimensions and increase efficiency of learning, we perform FPCA to decompose the time series.

```python
from skfda.exploratory.visualization import plot_fpca_perturbation_graphs
from skfda.preprocessing.dim_reduction.projection import FPCA
```

FPCA components:

```python
fpca_discretized_rw = FPCA(n_components=2)
fpca_discretized_rw.fit(fdgrx)
fpca_discretized_rw.components_.plot()
plt.xlabel('Timestep (hundreth of a sec)')
plt.ylabel('Roadwise coordinate (m)')
plt.grid(True)
plt.legend(['First FPCA component', 'Second FPCA component'])
print('')
```

FPCA mean:

```python
fpca_discretized_rw.mean_.plot()
plt.xlabel('Timestep (hundreth of a sec)')
plt.ylabel('Roadwise coordinate (m)')
plt.grid(True)
print('')
mean_curve = fpca_discretized_rw.mean_.data_matrix.reshape((-1))
```

Explained variance ratio

```python
fpca_discretized_rw.explained_variance_ratio_
```

Decompose signal into components: 

$ X(t) = \sum \xi_{ik} \phi_k(t) $ 

according to Karhunen-Loeve expansion. The $\phi_k$ are already known, but $\xi_{ik}$ must be calculated. It is easy:

$ \xi_{ik} = \int X(t) \phi_k(t)$

```python
friendly_basis_rw = fpca_discretized_rw.components_.data_matrix.squeeze().transpose()
coeffs_rw = np.matmul(px - mean_curve, friendly_basis_rw)
```

```python
plt.scatter(coeffs_rw[:,0], coeffs_rw[:,1])
plt.xlabel('First PCA component score')
plt.ylabel('Second PCA component score')
plt.grid(True)
```

It is very apparent that in this case, there is dominant cluster that represents the "normal traffic behavior"

## Distribution Fitting

We will fit a Gaussian mixture model onto the PCA-scores plot above. This is a hypothesis about dominant traffic mode in the patch, the "nominal behavior" hypothesis.

```python
from sklearn.mixture import GaussianMixture
```

```python
gm_rw = GaussianMixture(n_components=1, random_state=0).fit(coeffs_rw)
```

Plotting helper function

```python
import itertools
from scipy import linalg
import matplotlib as mpl
def plot_results(X, Y_, means, covariances, index, title):
    color_iter = itertools.cycle(['navy', 'c', 'cornflowerblue', 'gold',
                              'darkorange'])
    splot = plt.subplot(2, 1, 1 + index)
    for i, (mean, covar, color) in enumerate(zip(
            means, covariances, color_iter)):
        v, w = np.linalg.eigh(covar)
        v = 2. * np.sqrt(2.) * np.sqrt(v)
        u = w[0] / np.linalg.norm(w[0])
        # as the DP will not use every component it has access to
        # unless it needs it, we shouldn't plot the redundant
        # components.
        if not np.any(Y_ == i):
            continue
        plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1], .8, color=color)

        # Plot an ellipse to show the Gaussian component
        angle = np.arctan(u[1] / u[0])
        angle = 180. * angle / np.pi  # convert to degrees
        ell = mpl.patches.Ellipse(mean, v[0], v[1], 180. + angle, color=color)
        ell.set_clip_box(splot.bbox)
        ell.set_alpha(0.5)
        splot.add_artist(ell)

    plt.xlim(X[:,0].min(), X[:,0].max())
    plt.ylim(X[:,1].min(), X[:,1].max())
    plt.xticks(())
    plt.yticks(())
    plt.title(title)
```

```python
plot_results(coeffs_rw, gm_rw.predict(coeffs_rw), gm_rw.means_, gm_rw.covariances_, 0, 'Gaussian Mixture for Roadwise PCA components')
plt.xlabel('First PCA component score')
plt.ylabel('Second PCA component score')
plt.grid(True)
```

### Test Case 1: Whole abnormal trajectory

In this example, a speeding vehicle will be detected. The trajectory is whole. We present the conventional method of classification on the PCA-domain and time domain.

Load up the dangerous trajectory

```python
with open('vehicles_quitted_dangerous.pkl', 'rb') as input:
    dvehicles = pickle.load(input)
print('Data has been loaded successfully')
```

```python
plt.plot(dvehicles[0].tvec_i, dvehicles[0].log_pi[0,:])
```

Note that this trajectory is significantly shorter than nominal trajectories, indicating that this vehicle is overspeeding through the segment.

#### FPCA domain outlier detection
In the first case, we will see the option of outlier detection directly on FPCA domain

```python
pxd = np.array([v.log_pi[0] for v in dvehicles])
```

```python
coeffs_rwd = np.matmul(pxd - mean_curve, friendly_basis_rw)
```

```python
coeffs_rwd
```

```python
plt.scatter(coeffs_rw[:,0], coeffs_rw[:,1])
plt.scatter(coeffs_rwd[:,0], coeffs_rwd[:,1])
```

We clearly see that on FPCA domain, the outlier stays very far from the distribution. But the question is: can we detect this outlier as soon as it appears, so that we don't need to wait until it has exited the patch to get its representation on the FPCA components plane like this.

#### Time domain outlier detection

Let's give a name to our new shiny algorithm: **Outlier Trajectory Quickest Detector (OTQD)**

```python
%load_ext autoreload
%autoreload 2
```

```python
from otqd.otqd import OTQD
```

Initialize the OTQD module with $p_a$, parametrized by the mean $\mu_a$, information matrix $\Sigma^{-1}_a$, the noise $\sigma^{-2}_\epsilon$, the FPCA components, which is a matrix whose rows are time marks and columns are component values.

```python
otqd = OTQD(info_a = np.linalg.inv(gm_rw.covariances_),mu_a = gm_rw.means_.transpose(),info_e2 = 10.,pca_mean = mean_curve,pca_components = friendly_basis_rw)
```

We now add one measurement at time

```python
otqd.new_measurement(dvehicles[0].log_pi[0,0]) # first index is time, second index indicates that this is the x (roadwise) coordinate
```

```python
max_likelihood = np.max(otqd.calculate_log_likelihood())
```

We now consider gradually adding measurements over time

```python
ml_with_time_a = np.zeros((pxd.shape[1],))
otqd = OTQD(info_a = np.linalg.inv(gm_rw.covariances_),mu_a = gm_rw.means_.transpose(),info_e2 = 10.,pca_mean = mean_curve,pca_components = friendly_basis_rw)
for k in range(pxd.shape[1] - pxd.shape[1] + 30):
    otqd.new_measurement(pxd[0,k])
    ml_with_time_a[k] = otqd.calculate_log_likelihood()
```

In case of a nominal trajectory, we take the trajectory at the mean of the distribution. This is apparently the maximum likelihood of all trajectories

```python
pxr = mean_curve.reshape((-1,1)) + friendly_basis_rw @ (gm_rw.means_.transpose())
```

```python
ml_with_time_b = np.zeros((pxd.shape[1],))
otqd = OTQD(info_a = np.linalg.inv(gm_rw.covariances_),mu_a = gm_rw.means_.transpose(),info_e2 = 10.,pca_mean = mean_curve,pca_components = friendly_basis_rw)
for k in range(pxd.shape[1] - pxd.shape[1] + 30):
    otqd.new_measurement(pxr[k])
    ml_with_time_b[k] = otqd.calculate_log_likelihood()
```

```python
ml_with_time_c = np.zeros((px.shape[1],))
otqd = OTQD(info_a = np.linalg.inv(gm_rw.covariances_),mu_a = gm_rw.means_.transpose(),info_e2 = 10.,pca_mean = mean_curve,pca_components = friendly_basis_rw)
for k in range(px.shape[1] - px.shape[1] + 30):
    otqd.new_measurement(px[0][k])
    ml_with_time_c[k] = otqd.calculate_log_likelihood()
```

```python
fix, ax = plt.subplots()
time_for_plot = np.arange(10)/10.
plt.plot(time_for_plot, ml_with_time_c[0:10], label='Trajectory 1')
plt.plot(time_for_plot, ml_with_time_b[0:10], label='Nominal')
plt.plot(time_for_plot, ml_with_time_a[0:10], label='Outlier')
legend = ax.legend(loc='lower left', shadow=True)
plt.xlabel('Time since first enter to the patch (s)')
plt.ylabel('Log likelihood $\log \mathcal{L}_a$')
plt.grid(True)
```

We can try to plot all curves in the dataset, along with the outlier for better understanding

```python
fix, ax = plt.subplots()
time_for_plot = np.arange(10)/10.
for curve in range(px.shape[0]):
    ml_with_time_d = np.zeros((px.shape[1],))
    otqd = OTQD(info_a = np.linalg.inv(gm_rw.covariances_),mu_a = gm_rw.means_.transpose(),info_e2 = 10.,pca_mean = mean_curve,pca_components = friendly_basis_rw)
    for k in range(px.shape[1] - px.shape[1] + 30):
        otqd.new_measurement(px[curve][k])
        ml_with_time_d[k] = otqd.calculate_log_likelihood()
    plt.plot(time_for_plot, ml_with_time_d[0:10], label=str(curve))
plt.plot(time_for_plot, ml_with_time_a[0:10], label='Outlier')
plt.xlabel('Time since first enter to the patch (s)')
plt.ylabel('Log likelihood $\log \mathcal{L}_a$')
plt.grid(True)
```

It is evident that we can detect the outlier very rapidly. At just 6 timesteps, the difference (between the green curve and the nominal trajectory) has already been quite apparent.

### Test Case 2: Segmented abnormal trajectory
#### Time domain outlier detection with delay hypothesis

Suppose now we only observe a "segment of" measurements that spans over a period time $d < T$ and begins at $t_0 = $ 2s. Of course the algorithm does not know about this $t_0$.

```python
pxd_segment = pxd[0,20:40]
```

```python
plt.plot(np.arange(20)/10., pxd_segment)
plt.xlabel('Time since first observation (s)')
plt.ylabel('Roadwise coordinate (m)')
plt.grid(True)
```

```python
ml_with_time_d = np.zeros((pxd_segment.shape[0],))
delay_hypotheses = np.zeros((pxd_segment.shape[0],))
otqd = OTQD(info_a = np.linalg.inv(gm_rw.covariances_),mu_a = gm_rw.means_.transpose(),info_e2 = 10.,pca_mean = mean_curve,pca_components = friendly_basis_rw, i_max = 10)
for k in range(pxd_segment.shape[0]):
    otqd.new_measurement(pxd_segment[k])
    ll = otqd.calculate_log_likelihood()
    delay_hypotheses[k] = np.argmax(ll) 
    ml_with_time_d[k] = np.max(ll)
```

```python
fix, ax = plt.subplots()
time_for_plot = np.arange(ml_with_time_d[0:20].shape[0])/10.
plt.plot(time_for_plot, ml_with_time_d[0:20], label='Segment 1')
#plt.plot(ml_with_time_c[0:20], label='Trajectory 1')
plt.plot(time_for_plot, ml_with_time_b[0:20], label='Nominal')
plt.plot(time_for_plot, ml_with_time_a[0:20], label='Outlier')
legend = ax.legend(loc='lower left', shadow=True)
plt.xlabel('Time since first observation (s)')
plt.ylabel('Log likelihood $\log \mathcal{L}_a$')
plt.grid(True)
print('')
```

```python
plt.plot(delay_hypotheses)
```
