```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
x = np.arange(-1,1,0.01)
```

```python
y = np.zeros((x.shape[0]))
for k in range(1,100):
    if k % 2 == 0:
        y -= 2/((2*k-1)*np.pi) * np.cos((2*k-1) * np.pi/2 * x)
    else:
        y += 2/((2*k-1)*np.pi) * np.cos((2*k-1) * np.pi/2 * x)
plt.plot(x,y)
```
