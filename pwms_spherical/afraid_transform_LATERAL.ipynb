# Analytical decomposition of the Wiener Process

Hopefully we can retrieve the AFRAID Transform.

```python
import numpy as np
from matplotlib import pyplot as plt 

traj_length = 500
```

```python
t_vec = np.arange(traj_length)/traj_length
wp_basis = np.zeros((traj_length, traj_length))
for k in range(traj_length):
    wp_basis[k,:] = np.sqrt(2) * np.sin((k+0.5)*np.pi*t_vec)
    wp_basis[k,:] = wp_basis[k,:] / np.linalg.norm(wp_basis[k,:])
    if k<5:
        plt.plot(wp_basis[k,:])
```

```python
import pickle 
with open('frsid_'+str(traj_length)+'.basis', 'wb') as f:
    pickle.dump(wp_basis.T, f)
print('FRSID Transform of length {:d} has been generated successfully!'.format(traj_length))
```
