```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
x = np.arange(0,1.01,0.01)
basis = np.zeros((3, x.shape[0]))
for k in range(1, 4):
    basis[k-1,:] = np.cos((k-1/2)*np.pi*x) - 1
    plt.plot(basis[k-1,:])
```

```python
np.sum(basis[1,:] * basis[1,:])/100
```
