```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
alpha_set = [0.1, 0.3, 0.5, 0.7, 0.9, 1.2]
for alpha in alpha_set:
    plt.figure()
    cov = np.zeros((500,500))
    for i in range(cov.shape[0]):
        for j in range(cov.shape[1]):
            cov[i,j] = np.exp(-alpha * (i+j)/500) * 1/alpha * (np.exp(alpha * np.min((i,j))/500 ) - 1)
    plt.contourf(cov)
    plt.axis('equal')
    plt.xlabel('s')
    plt.ylabel('t')
    plt.gca().invert_yaxis()
    plt.title(r'$\alpha$ = {}'.format(alpha))
    plt.savefig('xcovfrsid_alpha_{}.pdf'.format(alpha), bbox_inches='tight')
```

```python
alpha_set = [0.1, 0.3, 0.5, 0.7, 0.9, 1.2]
for alpha in alpha_set:
    plt.figure()
    cov = np.zeros((500,500))
    for i in range(cov.shape[0]):
        for j in range(cov.shape[1]):
            cov[i,j] = np.exp(-alpha * (i+j)/500) * 1/alpha * (np.exp(alpha * np.min((i,j))/500 ) - 1)
    eigval, eigvec = np.linalg.eig(cov)
    for k in range(5):
        plt.plot(eigvec[:,k], label=str(k+1))
    plt.ylabel('t')
    plt.grid(True)
    plt.title(r'$\alpha$ = {}'.format(alpha))
    plt.savefig('kltx_{}.pdf'.format(alpha), bbox_inches='tight')
```
